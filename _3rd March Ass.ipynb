```python
#Answer-1
#List of five plot that we can plot using seaborn library
# 1- Scatter plot->
# A scatter plot is a visual representation of how two variables relate to each other. You can use scatter plots to explore the relationship between two variables, for example by looking for any correlation between them.


# 2- Bar plot -> 
# A barplot is basically used to aggregate the categorical data according to some methods and by default it’s the mean

# 3- Distplot->
# We use the seaborn dist plots to plot histograms with the given variables and data as a result.


# 4-Countplot
# A countplot basically counts the categories and returns a count of their occurrences.
# It is one of the simplest plots provided by the seaborn library.

# 5-Line Plot
# We use the seaborn line plot mainly to visualize the given data in some time-series form, i.e., in a continuous manner with respect to time.
```

```python
#Answer-2
import seaborn as sns 
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.lineplot(x = "timepoint",y = "signal", data = fmri , hue = "event",style= "region")
```

```python
#Answer-3
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.boxplot(x = "pclass",y = "age" , data = titanic)
```

```python
sns.boxplot(x = "pclass",y = "fare" , data = titanic)
```

```python
#Answer-4
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
sns.displot(x = "price",data = diamonds,hue = "cut")
```

```python
#Answer-5
iris = sns.load_dataset("iris")
iris.tail()
```

```python
sns.pairplot(iris,hue = "species")
```
